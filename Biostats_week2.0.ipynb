```r
pVals<-c()
wVals<-c()
```

```r
set.seed(19)
```

```r
z=rpois(5000, lambda=10)
```

```r
for (i in 1:5000)
{
  #add conditional statement here
    #if(i%in%c(1:50)){
       # z=z+rexp(5000)
    #}
        #else {
           # z=z+rpois(5000, lambda=50)
       # }
  #if the value is between 1 and 50, add a exponential distr
  #else if the value is between 60 and 100, add a poison dist
  z = z+rpois(5000, lambda=10)
    #runif(5000)     #add another uniformly distributed variable
  t = shapiro.test(z)
  pVals[i] = t$p.value
  wVals[i] = t$statistic
}
```

```r
hist(z)
```

```r
plot(c(1:5000),pVals,type="b",xlab='Iteration')
abline(h=0.05)
#plot(c(1:1000),wVals,type="b",xlab='Iteration')
```
