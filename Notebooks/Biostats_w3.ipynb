```r
# A t-test is a parametric test - certain assumptions need to be met. Normal dist
single_sample<-rnorm(10, mean=10)
t.test(single_sample, mu=10)
```

```r
single_sample = rnorm(100,mean=100)# using a mean of 100
t.test(single_sample, mu = 10)                #test if average=10
```

```r
#Two sample t-test ----

x = rnorm(10)
y = rnorm(10)
boxplot(x,y,names=c("X","Y"))

t.test(x,y)
#Null hypothesis: True difference in means is equal to 0
```

```r
#un-paired two sample t-test

x = rnorm(10)
y = rnorm(40)
boxplot(x,y,names=c("X","Y"))
boxplot(x,y,names=c("X","Y"),outline=F)

t.test(x,y)
#Null hypothesis: True difference in means is equal to 0
```
